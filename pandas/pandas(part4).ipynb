```python
import pandas as pd 
import matplotlib.pyplot as plt 
 
```

```python
# 선 그래프

df = pd.read_excel('./part4/inandout.xlsx',fillna=0,header=0)
# filna()메소드의 method =ffill을 사용하면 누락 데이터 행 바로 앞에 위치한 행의 데이터 값으로 채워준다

df =df.fillna(method='ffill')# 누락 데이터를 앞 에이터로 채움(엑셀 양식 병합 부분)

#서울에서 다른 지역으로 이동한 데이터만 추출 정리
mask = (df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')
df_seoul=df[mask]
df_seoul=df_seoul.drop(['전출지별'],axis=1)
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)
df_seoul.set_index('전입지',inplace=True)
print(df_seoul.head())
```

```python
#서울에서 경기ㅏ도로 이동한 인구 데이터만 선택
sr_one = df_seoul.loc['경기도']
print(sr_one.head())
```

```python
plt.plot(sr_one.index,sr_one.values)
```

```python
# 그래프 꾸미기 
plt.figure(figsize=(14,5))#크기지정

plt.xticks(rotation='vertical')#x-label 회전\


plt.plot(sr_one)#series or dataframe객체 자체도 입력 가능 결과는 같다 

plt.title('seoul move to gungki',size=50)

plt.xlabel('period',size=30)

plt.ylabel('moving count',size=30)  

plt.legend(labels=['seoul->gungki'],loc='best',fontsize=15) #범례 표시

plt.show()
```

```python
#스타일 서식 지정
plt.style.use('ggplot') # 스타일 서식 지정

plt.figure(figsize=(19,5))


plt.xticks(rotation='vertical')#x-label 회전\
 

plt.title('seoul move to gungki',size=50)

plt.xlabel('period',size=30)

plt.ylabel('moving count',size=30)  

plt.legend(labels=['seoul->gungki'],loc='best',fontsize=15) #범례 표시

plt.plot(sr_one,marker='o',markersize=10)# 마커 표시

plt.show()


# marker = 'O'
# markerfacecolor  마커 색
# markersize 
#color 선 색
#linewidth 선 두께
#label
```
