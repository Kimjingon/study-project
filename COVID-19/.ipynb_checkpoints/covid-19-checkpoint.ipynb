<h1> 라이브러리 </h1>

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import seaborn as sns 
from scipy.stats import pearsonr,spearmanr # 선형적 상관관계, 관계 여부 판단
from datetime import datetime, timedelta
```

<h1> 그래프 스타일 지정 </h1>

```python
plt.style.use('dark_background')
mpl.rcParams['axes.unicode_minus']=False 
```

<h1> 전처리 </h1>

```python
#전처리
pos_questions= [
    'OPN1','OPN3','OPN5','OPN7','OPN8','OPN9','OPN10', #Openness 개방성
    'CSN1','CSN3','CSN5','CSN7','CSN9','CSN10', # 성실성
    'EXT1','EXT3','EXT5','EXT7','EXT9', #외향성 
    'AGR2','AGR4','AGR6','AGR8','AGR9','AGR10', #친화성
    'EST1','EST3','EST5','EST6','EST7','EST8','EST9','EST10', #안정성
]
neg_questions =[
    'OPN2','OPN4','OPN6', #Openness 개방성
    'CSN2','CSN4','CSN6','CSN8', # 성실성
    'EXT2','EXT4','EXT6','EXT8','EXT10', #외향성 
    'AGR1','AGR3','AGR5','AGR7', #친화성
    'EST2','EST4', #안정성
]

usecols = pos_questions + neg_questions + ['country']# column 지정
```

```python
#지정한 column의 데이터만 가져오기
df=pd.read_csv('./IPIP-FFM-data-8Nov2018/data-final.csv',sep='\t', usecols=usecols)
print(df.head())
print("\n")
print(len(df))
```

```python
df=df.replace(0,np.nan).dropna(axis=0).reset_index(drop=True)
#0인 값은 누락 데이터로 처리해서 drop처리

print(len(df))
```

```python
df_2 = (df.groupby('country').agg('count')['EXT1']>100).reset_index()
#나라를 기준으로 평균을 낼 건데, 답변이 너무 적으면 특성 나타내기 힘듦
# 100명 이상 답변한 나라만 추려냄
print(df_2)
print("\n")

fc = df_2[df_2['EXT1']==True]['country']
df = df[df['country'].isin(fc)].reset_index(drop=True)
print(df.head())

```

<h1> 데이터  값 정규화 </h1>

```python
#-2~2
df[pos_questions]=df[pos_questions].replace({1:-2,2:-1,3:0,4:1,5:2})
df[neg_questions]=df[neg_questions].replace({1:2,2:1,3:0,4:-1,5:-2})# neg이므로 반대로
df.head()
```

<h1> Compute Scores(흩어진 항목을 5개 항목으로) </h1>

```python
traits={
    'OPN':'개방성',
    'CSN':'성실성',
    'EXT':'외형성',
    'AGR':'친화성',
    'EST':'안정성'
}


for trait,trait_kor in traits.items():
    trait_cols = sorted([col for col in df.columns if trait in col])
    df[trait_kor]=df[trait_cols].sum(axis=1)
    print(trait_cols)

    #5개 항목을 돌면서 col안에 있는 trait이면 trait_cols에 저장 후 다 더해서 df[trait_kor]에 저장

df_traits = df[list(traits.values())+['country']].copy()
df_traits.head()
```

<h1>Distribution Plot</h1>

```python
fig = plt.figure(figsize=(16,6))
for trait in traits.values():
    sns.distplot(df_traits[trait], kde =True, bins=40, axlabel=False)
#kde =false 로 하면 밀도 그래프는 그리지 않는다
#bins = 막대의 너비를 결정 
    fig.legend(traits.values())
    
```

<h1>Opnness Korea vs Germany</h1>

```python
fig = plt.figure(figsize=(16,6))# 그림사진
sns.distplot(df_traits[df_traits['country']=='KR']['개방성'],bins=40, axlabel=False)
sns.distplot(df_traits[df_traits['country']=='DE']['개방성'],bins=40, axlabel=False)
fig.legend(['KR','DE'])

#개방성 관련해서는 독일의 점수가 더 높게 분포된 것을 알 수 있다.
```

<h1> 나라별 평균값 계산</h1>

```python
df_traits_mean=df_traits.groupby('country').mean().rename_axis('country').reset_index()

df_traits_mean

#column 이름 country로 바꾸고 index로 country 지정
```

<h1>COVID-19 Dataset</h1>

```python
df_covid =pd.read_csv('https://raw.githubusercontent.com/datasets/covid-19/master/data/time-series-19-covid-combined.csv',parse_dates=['Date'])
#parsed_date 를 통해 Date를 date 형식으로 parsing 한다.
df_covid.head() 
```

<h1> Country Code</h1>

```python
cc =pd.read_csv('code.csv')
cc.head()
# 두개를 join시켜줘야함 
```

<h1>Filter Dataset step 1</h1>

```python
df_covid = df_covid[df_covid['Confirmed']>50].reset_index(drop=True)
#50명 이상이여아한다.
df_covid = df_covid.groupby(['Country/Region','Date']).sum().reset_index()
#주나 도 단위는 나라로 묶어주기
df_covid[df_covid['Country/Region']=='US']
```

<h1>Filter Dataset step 2</h1>

```python
n_days=14
filtered=(
    datetime.now()-df_covid.groupby('Country/Region')['Date'].min()>timedelta(days=n_days)
).reset_index().rename(columns={'Date':'Filtered'})
#지금 날짜 - 50명이 된 날짜가 14일 이상 차이되는 것 filtered 열에 true,false로 입력

print(filtered)
print("\n")
filtered_countries = filtered[filtered['Filtered']==True]['Country/Region']
#true 인 나라만 입력


print(filtered_countries)
print("\n")


df_covid = df_covid[df_covid['Country/Region'].isin(filtered_countries)]
#filtered 된 나라의 dataframe

df_covid_14days = df_covid.groupby('Country/Region').head(n_days).groupby('Country/Region').tail(1)
# 첫 14일간 데이터에서 tail(1)마지막 날 데이터 
print(df_covid_14days)
 
```

<h1>Merge All</h1>

```python
df_covid_14days=df_covid_14days.merge(cc,left_on='Country/Region',right_on='Name')

df_covid_14days = df_covid_14days.merge(df_traits_mean,left_on='Code',right_on='country')
df_covid_14days.sort_values('Confirmed',ascending=False)

```

<h1>Compute Pearson Correation</h1>

```python
#선형적 관계 분석 
#out lined 제거 
#china ,turkey 너무 비정상 적으로 큼 

new_df =df_covid_14days[~df_covid_14days['country'].isin(['CN','TR'])]

for trait, trait_kor in traits.items():
    fig, axes = plt.subplots(1,3,figsize=(20,6))
    
    for i,n in enumerate(['Confirmed','Recovered','Deaths']):
        corr = pearsonr(
            new_df[trait_kor],
            new_df[n]
        )
        
        sns.regplot(x=trait_kor, y= n , data=new_df, ax=axes[i])
        axes[i].set_title('%s,%s ::r=%.2f, p=%.2f'% (trait_kor,n,corr[0],corr[1]))
        
        #corr[0] (Pearson coefficient) : 선형적 상관관계가 얼마나 큰지
        #corr[1] (P-value) :틀릴 확률
        
```
